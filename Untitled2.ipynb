```python

# Integer
a = 10
print("Integer value:", a)
print("Type:", type(a))
# Float
b = 10.5
print("\nFloat value:", b)
print("Type:", type(b))
# String
c = "Hello, Python"
print("\nString value:", c)
print("Type:", type(c))
# Boolean
d = True
print("\nBoolean value:", d)
print("Type:", type(d))
# List (mutable)
e = [1, 2, 3, "apple"]
print("\nList value:", e)
print("Type:", type(e))
# Tuple (immutable)
f = (1, 2, 3, "banana")
print("\nTuple value:", f)
print("Type:", type(f))
# Dictionary (key-value pairs)
g = {"name": "Pallavi", "age": 18}
print("\nDictionary value:", g)
print("Type:", type(g))
# Set (unique values)
h = {1, 2, 3, 3, 4}
print("\nSet value:", h)
print("Type:", type(h))
```
