```python


# Creating a tuple
tuple1 = (1, 2, 3, 4, 5)
tuple2 = (6, 7, 8)

# 1. Length of tuple
print("Tuple1:", tuple1)
print("Length of tuple1:", len(tuple1))

# 2. Concatenation
concat_tuple = tuple1 + tuple2
print("Concatenated tuple:", concat_tuple)

# 3. Repetition
repeat_tuple = tuple1 * 2
print("Repeated tuple:", repeat_tuple)

# 4. Membership
element = 3
if element in tuple1:
    print(element, "is present in tuple1")
else:
    print(element, "is not present in tuple1")
```
